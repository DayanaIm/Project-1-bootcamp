```python
# Assuming your DataFrame is named 'df'

# Group the DataFrame by 'year'
grouped = primary_data.groupby('year')

# Aggregate the data to calculate the total of 'sex' and 'age' for each year
result = grouped.agg({
    'sex': 'count',  # Count the number of rows for each year
    'age': 'sum'     # Sum the 'age' values for each year (assuming 'age' is numeric)
}).reset_index()

# Rename the columns if needed
result = result.rename(columns={'sex': 'total_sex', 'age': 'total_age'})

# Display the resulting DataFrame
print(result)
```
